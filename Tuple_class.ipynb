## Unpacking method
#### Tuple unpacking means assigning the values of a tuple to multiple variables in one line.

```python
tuple3=(1,2,3,"moss-tech","vietnam")
(a,b,c,d,e)=tuple3
print(a,b,c,d,e)
```

```python
tuple4=(1,2,3,"moss-tech","vietnam")
a,b,c,d,e=tuple4 # can use without ()
print(a,b,c,d,e)
```

```python
tuple5=(1,2,3,"moss-tech","vietnam")
a,b=tuple5
```

```python
# * is called as krgs one star is used for single items not for key value pairs
# ** is called kwrgs
a,*b,*c=tuple5  
print(a,b,c)
```

```python
a,b,*c=tuple5  
print(a,b,c)
```

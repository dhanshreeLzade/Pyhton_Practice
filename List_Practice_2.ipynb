```python
#Remove duplicates from a list while maintaining order
list6=[1,2,32,1,2,6]
unique_list=[]
for item in list6:
    if item  not in unique_list:
        unique_list.append(item)
print(unique_list)
```

```python
#Remove duplicates from a list while maintaining order
list7=[1,2,32,1,2,6]
list7=list(set(list7))
print(list7)
```

```python
list6 = [1, 2, 32, 1, 2, 6]  # Original list with duplicates

# Loop over indices in reverse order (from last index to first)
for i in reversed(range(len(list6))):
    # Check if current element appears **earlier in the list**
    if list6[i] in list6[:i]:
        # If it does, delete it from the list
        del list6[i]

# Print the modified list (duplicates removed, order maintained)
print(list6)  # Output: [1, 2, 32, 6]
```

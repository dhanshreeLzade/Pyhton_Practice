```python
‚≠ê Flow of Execution in Functions (Python)

Flow of execution means the order in which Python runs the statements when functions are involved.

‚úî 1. Python starts at the top of the program

It reads code line-by-line.

‚úî 2. When Python sees a function call, it JUMPS to the function

It stops main code ‚Üí goes inside the function ‚Üí runs all lines inside that function.

‚úî 3. After finishing the function, Python returns back

It comes back to the line after the function call.

‚≠ê Simple Example
def greet():               # Step 1: Python reads function definition
    print("Hello!")

print("Start")             # Step 2: Runs this
greet()                    # Step 3: Function call ‚Üí jump inside greet()
print("End")               # Step 4: Comes back and prints this

Flow

Python reads function definition (does NOT run it yet)

Prints Start

Sees greet() ‚Üí jumps into the function

Prints Hello!

Returns to the main code

Prints End
```

```python
Function Definition in Interactive Mode (Python)
Interactive mode means writing Python code directly in the Python shell / interpreter (like IDLE, terminal, or Jupyter).
You can define a function directly there by typing it line-by-line.
Steps:

Type def function_name():

Press Enter ‚Üí Python shows ... (means you are inside the function)

Write function body (statements)

Press Enter on an empty line to finish the function

Now you can call the function normally

‚≠ê Why special?

Because in interactive mode:

Python takes input line-by-line

You must leave one blank line to end the function definition

Then the function is ready to use immediately
```

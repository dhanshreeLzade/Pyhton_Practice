```python
class c:
    def __init__(self,mas,mini,lst):
        try:
            for i in lst:
                if i>mas:
                    mas=i
                    print(mas)
                elif i<mini:
                    mini=i
                    print(mini)
            return mini
        except Exception as e:
            print("wrong input pass")
obj=c(0,99999999999,[11,1,2,2,3,3223,21])
```

```python
class c:
    def __init__(self, mas, mini, lst):
        try:
            for i in lst:
                if i > mas:
                    mas = i
                    # print(mas)   ‚ùå remove this
                    self.mas = mas      # ‚úÖ store in object
                elif i < mini:
                    mini = i
                    # print(mini)  ‚ùå remove this
                    self.mini = mini    # ‚úÖ store in object

            print("Max:", self.mas)
            print("Min:", self.mini)
        except Exception as e:
            print("wrong input pass")
obj=c(0,99999999999,[11,1,2,2,3,3223,21])
```

```python
# Remove duplicates from a tuple.
t=(3, 1, 2,2,2,33,0,3,3, 4, 5, 6, 7, 8)
unique_value=[]
for i in t:
    if i  not in unique_value:
        unique_value.append(i)
print(tuple(unique_value))
```

```python
âœ… Tuples in Python are immutable

You cannot change a tuple after creating it.

You cannot add, remove, or update items inside a tuple.

So duplicates cannot be removed inside the same tuple.
```

```python
#Merge multiple tuples into one.
t=(3, 1, 2, 33, 0, 4, 5, 6, 7, 8)
p=(0,0,0)
print(t+p)
merged = sum((t, p), ())
print(merged)
```

```python
#Count the frequency of all elements in a tuple.
#Frequency means â†’ how many times each element appears.
t=(3, 3,1, 2, 33, 0, 4, 5, 6, 7, 8)
for target in t:
    count=0
    
    for i in t:
        if i == target:
            count=count+1
        
    print(target,"number occurs this much times-",count) #this should uder for else see below what happeened..................
```

```python
t = (3, 3, 1, 2, 33, 0, 4, 5, 6, 7, 8)

for target in t:        # take each element as target
    count = 0           # reset count every time

    for i in t:         # check how many times target appears
        if i == target:
            count = count + 1

    print(target, "occurs", count, "times")
```

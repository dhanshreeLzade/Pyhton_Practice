# Remove methods
### remove()	Removes a specific element; error if not found	s.remove(2)
### discard()	Removes a specific element; no error if not found	s.discard(2)
### pop()	Removes and returns a random element	s.pop()
### clear()	Removes all elements	s.clear()

```python
set_1={1,2,3,4,5,6,67}
print(set_1)
set_1.add(4555556)
set_1.remove(6) #set.remove() takes exactly one argument (2 given)
print(set_1)
set_1.discard(100)#set.discard() takes exactly one argument
print(set_1)
set_1.pop() #set.pop() takes no arguments (1 given)
print(set_1)
```

```python
set_2={1,2,3,4,5,6,67}
set_2.clear()#set.clear() takes no arguments (1 given)
set_2
```

# SET OPERATION + _|*

```python
1️⃣ Union (| or union())
```

```python
set_3={1,2,3,4,5,6,67} # no change
set_4={"marry",45} # no change
set_3.union(set_4)#method doesnt update any sets just give new set-------
set5={090.23}
```

```python
set_3|set_4|set5
```

```python
Intersection (& or intersection())
```

```python
set_5={1,2,3,4,5,6,67} # no change
set_6={"marry",45,1,2} # no change
set_5.intersection(set_6) #Common items dete hi method
```

```python
set_5 & set_6 #intersection operation
```

3️⃣ Difference (- or difference())
Keeps elements that are in the first set but not in the second.

```python
set_5={1,2,3,4,5,6,67} # no chan
set_6={"marry",45,1,2} # no change
set_5.difference(set_6)#coomon vale delete krte ani je pahila set ahe tyache purn value ghete
```

```python
set_5 - set_6
```

```python
4️⃣ Symmetric Difference (^ or symmetric_difference())
Keeps elements that are in either set, but not in both.
```

```python
set_5={1,2,3,4,5,6,67} # no chan
set_6={"marry",45,1,2} # no change
set_5.symmetric_difference(set_6)#coomon vale delete krte ani donhi set che ahe tyache purn value ghete
```

```python
set_5 ^ set_6
```

```python
5️⃣ Membership (in / not in)

Check if an element exists in a set.
```

```python
set_5={1,2,3,4,5,6,67} # no chan
set_6={"marry",45,1,2} # no change
0 not in set_5
```

```python
isdisjoint()	Checks if two sets have no common elements	s1.isdisjoint(s2) → True/False
issubset()	Checks if all elements of one set are in another	s1.issubset(s2) → True/False
issuperset()	Checks if a set contains all elements of another	s1.issuperset(s2) → True/False
```

```python
set_5={1,2,3,4,5,6,67} # no chan
set_6={"marry",45,1,2} # no change
set_5.isdisjoint(set_6) #common nahi pahije
```

```python
set_5={1,2,3,4,5,6,6} # sett
set_6={1,2,3,1} #subset
set_5.issubset(set_6)
```

```python
set_5={1,2,342,32,232424,} #supersett
set_6={1,2,3} #subset
set_5.issuperset(set_6)
```

```python
difference_update(other_set)	Removes elements found in another set	s1.difference_update(s2)
symmetric_difference_update(other_set)	Keeps elements in either set but not both	s1.symmetric_difference_update(s2)
```

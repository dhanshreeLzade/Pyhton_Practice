```python
s = "my mail is aishwaryazade24@gmail.com please note"

i = 0
email = ""        # not a loop variable, just collecting characters
start = False     # tells when email started

while i < len(s):

    if s[i] == "@":          # email definitely starts here
        start = True

        # go backward using i to find the first letter of email
        while s[i] != " ":
            i -= 1
        i += 1               # move to first email character

        # now print forward full email
        while i < len(s) and s[i] != " ":
            email += s[i]
            i += 1
        break

    i += 1

print(email)
```

```python
# Input
name1 = input()
dob1 = input()   # DD-MM-YYYY

name2 = input()
dob2 = input()   # DD-MM-YYYY

# Split DOB using while loop
def split_dob(d):
    part = ["", "", ""]
    i = 0
    p = 0
    
    while i < len(d):
        if d[i] == "-":
            p += 1
        else:
            part[p] += d[i]
        i += 1
    
    # convert to integer
    return int(part[0]), int(part[1]), int(part[2])   # day, month, year

d1, m1, y1 = split_dob(dob1)
d2, m2, y2 = split_dob(dob2)

# Comparison logic
# Younger person → LATER date (greater year/month/day)

younger = ""

# Compare year
if y1 > y2:
    younger = name1
elif y2 > y1:
    younger = name2
else:
    # Year equal → compare month
    if m1 > m2:
        younger = name1
    elif m2 > m1:
        younger = name2
    else:
        # Month equal → compare day
        if d1 > d2:
            younger = name1
        elif d2 > d1:
            younger = name2
        else:
            # Same DOB → alphabetical order
            if name1 < name2:
                younger = name1
            else:
                younger = name2

print(younger)
```

```python
sample_list = [87, 45, 41, 65, 94, 41, 99, 94]

# ---------- Remove duplicates ----------
unique = []

i = 0
while i < len(sample_list):
    if sample_list[i] not in unique:
        unique.append(sample_list[i])
    i = i + 1

print("unique items", unique)

# ---------- Convert to tuple ----------
t = tuple(unique)
print("tuple", t)

# ---------- Find min & max using while loop ----------

# initialize
min_val = t[0]
max_val = t[0]

j = 0
while j < len(t):
    if t[j] < min_val:
        min_val = t[j]
    if t[j] > max_val:
        max_val = t[j]
    j = j + 1

print("min:", min_val)
print("max:", max_val)
```

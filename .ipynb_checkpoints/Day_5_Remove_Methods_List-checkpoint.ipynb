# Using POP()-

```python
Listnum= [1,"hello",
    ["apple", "banana", [100, 200, 300]],
    42,
    ["cat", ["dog", ["parrot", 555, "lion"], "tiger"], "cow"],
    "end",
    [10, 20, 30],
    ["deep", ["deeper", ["deepest", 999]]]
]
print(Listnum)
```

```python
Listnum.pop(3)
print(Listnum)
```

```python
Listnum[3][1].pop(1)
print(Listnum)
```

```python
num= [1,"hello",
    ["apple", "banana", [100, 200, 300]],
    42,
    ["cat", ["dog", ["parrot", 555, "lion"], "tiger"], "cow"],
    "end",
    [10, 20, 30],
    ["deep", ["deeper", ["deepest", 999]]]
]
print(num)
```

```python
num[4][1].pop(2)
print(num)
```

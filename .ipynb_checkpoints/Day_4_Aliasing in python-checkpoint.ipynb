# Aliasing
#### *Aliasing happens in python and two variable are referencing to the same object in memory or pointing towards the same object in the memory*
#### E.g - a=[1,2,3,4,5]
####       b=a
#### Note- Aliasing can happens with muttable elements like list , set , dictonary and not with int string as they are not muttable.

```python
a=[1,2,3,4]
b=a
print(b)
```

```python
b.append(5)
```

#### if we change one variable then change will also reflect in other variable too..  

```python
print(a)
```

```python
a.append(1)
print(a)
```

#### here we learnt when we use **append** the item will get inserted from the end and not from in begining or in middle....

```python
#Now
list1=[1,2,3,4,5]
list2=[1,2,3,4,5]
## as above both variables have same values but is it aliasing?
## lets check through one method
print (list1 is list2)
```

```python
#### it is Giving false because  list1 and list2 is created separately even though values are same
#### but they are referencing too different objects in memory
```

```python
#### if we want to check values are same at that time we use below method
list1 == list2
```

```python
#### it is giving true cause values are same...
```

```python
list=[1,2,3,4,5]
list2=list
list2 is list
```

```python
# now its is giving True.......cause they are referencing too same objects in memory
```

## To avoid aliasing what to do ??

#### *copy() - it is used when we want copy values from single list whereas* 
####  *deepcopy() is used when we have to copy values from nested list*

```python
leest=[1,2,3,3,4,4,"diwali" ,2.3]
leest2= leest.copy()
leest2
```

```python
leest2.append("jio hotstar")
```

```python
leest2
```

```python
leest
```

```python
#here we can see leest is not affected when we use append method because we use copy
```

## String to complex

```python
variable6="19"
print (type(variable6))
variable6= complex(variable6)
print (variable6 ,type(variable6))
print (variable6.real)
print (variable6.imag)
```

```python
## here 19 is real
## 0j is imaginory part
#part before + is real and part after + is imag
## real and imag are attributes 
""" In python when we give . after value or variable its usually a attrribute or method
Attribute = a value or property that an object has
(like a characteristic or feature)
Example: car.color — the car has a color

Method = a function or action that an object can do
Example: car.drive() — the car can drive

we use parenthesis fof method not for object 
"""
```

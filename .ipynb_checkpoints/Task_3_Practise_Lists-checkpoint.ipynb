# Find common elements between two lists
### This is a set intersection in Python — used to find common elements between two lists.

```python
c=[1,2,3,4,5,6,7,8]
d=[2,3,"aishwarya","aishika",2.3,2.4]
common=list(set(c) & set(d))
```

```python
print(common)
```

# Check if a list is a palindrome -- it is unchanged when reverse e.g madam
## check reveerse compare list

```python
a=["madam","mom","madam"]
if a == a[::-1] :
 print("it is palindrome") 
else :
 print("its is not")
```

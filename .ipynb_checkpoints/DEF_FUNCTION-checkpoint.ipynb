```python
def calsi():
    user_inp1=int(input("Enter the number"))
    user_inp2=int(input("Enter the number"))
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    user_choice=int(input("Enter operartion you want to perform :"))
    if user_choice==1:
        add=user_inp1+user_inp2
        print("1. Addition",add)
    elif user_choice==2:
        sub=user_inp1-user_inp2
        print("2. Subtraction",sub)
    elif user_choice==3:
        mul=user_inp1*user_inp2
        print("3. Multiplication",mul)
    elif user_choice==4:
         div=user_inp1/user_inp2
         print("4. Division",div)
    else:
        print("Invalid choice")

calsi() #function call is important
```

```python
# Write a function that takes two numbers and prints the greater number.
def num(a,b):
    if a>b:
        print(a,"max")
    else:
        print(b,"max")
```

```python
# Write a function that takes number and divisor and prints whether it is divisible.
def divi(num,divis):
    if num%divis==0:
        print(f"{num}it is divible by {divis}")
    else:
        print("not didvisible")
divi(10,5)
```

```python
Lists & Strings as parameters

Write a function that takes a list and prints sum of elements.

Write a function that takes a list and prints largest element.

Write a function that takes a list and removes duplicates.

Write a function that takes a string and prints vowel count.

Write a function that takes a string and checks palindrome.
```

```python
# Write a function that takes a list and prints sum of elements.
def fun(lst):
    sum=0
    for i in lst:
        sum=sum+i
    print(sum)
fun([1,2,3])
```

```python
# Write a function that takes a list and prints largest element.
def fun(lst):
    num=0
    for i in lst:
        if i>num:
            num=i
    print(f"greater is : {num}")
fun([2,21231,223,322133,212,0])
```

```python
# Write a function that takes a list and removes duplicates.
def nm(lst):
    lst1=[]
    for i in lst:
        if i not in lst1:
            lst1.append(i)
        else:
            lst.pop(i)
            print(lst)
nm([232,22,3,3,2,13])
```

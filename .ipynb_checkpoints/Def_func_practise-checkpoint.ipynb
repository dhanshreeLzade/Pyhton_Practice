```python
# Write a function to reverse a string
def rev():
    stri="heena"
    rev=""
    for i in stri:
        rev=i+rev
    print(rev)
rev()
```

```python
# Write a function to print count of vowels in a string
def c():
    vowels = ('a','e','i','o','u','A','E','I','O','U')
    stri = "Ananya parlawar"
    count = 0

    for i in stri:
        if i in vowels:
            count = count + 1

    print("Vowel count:", count)

c()
```

```python
# Write a function that prints factorial of a number.
def fun():
    fact=1
    user_inp=int(input("enter the number : "))
    for i in range(1, user_inp+1):
        fact=fact*i
        print(fact)
fun()
```

```python
#Write a function to print Fibonacci series up to n terms.
# Write a function to print Fibonacci series
def fibo():
    n = int(input("Enter how many terms: "))

    a = 0
    b = 1

    for i in range(n):
        print(a)
        c = a + b
        a = b
        b = c

fibo()
```

```python
# Write a function that prints maximum number in a list.
def maxim():
    list1=[1,2,3,4,5,3,4,78,7,787,300]
    max=0
    for i in list1:
        if i >max:
           max=i
    print(max,f"max is {max}")
maxim()        
```

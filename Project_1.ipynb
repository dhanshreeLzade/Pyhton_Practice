# Password Strength Checker

```python
password=input("Enter the password: ")
alpha=False
digit=False
spe=False
if len(password)<8:
        print("Password should be alteast more than eight characters")
for ch in password:
        if('a'<= ch <='z') or ('A'<= ch <='Z'):
            alpha=True
        elif ('0' <=ch <='9'):
            digit=True
        else:
            spe=True
if alpha==True and digit==True and spe==True:
    print("Stroger password")
else:
    print("weaker")
```

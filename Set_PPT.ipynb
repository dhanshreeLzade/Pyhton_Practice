# 5. Subset (A ⊆ B) → issubset()
### 1. issubset()  set1.issubset(set2)
#### → All elements of set1 must be present in set2.
#### → Even one element missing = False.

```python
set1={1,2,3,4,67}
set2={5,6,2,7}
set1.issubset(set2)
set1.issuperset(set2)
```

```python
set1={1,2,3,4,67}
set2={1,2,0}
set1.issubset(set2)
set1.issuperset(set2)
```

```python
set1={1,2,3,4,67}
set2={1,2,1}#here element should in firstset
set1.issuperset(set2)
```

## 6 .Disjoint → isdisjoint()
## Checks if sets have no common elements.

```python
set1={1,2,3,4,67}
set2={1,2,1}# common elements are there.........
set1.isdisjoint(set2)
```

```python
set1={1,2,3,4,67}
set2={7,9}# no common elements are there.........
set1.isdisjoint(set2)
```

# intersection() → gives new set (does not change original set).
# intersection_update() → updates the original set and keeps only common items.


```python
set1={1,2,3,4,5,6}
set2={7,8,2,3,4}
set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5,6,2,3,11} #common from second set and remaining from sel1
set2={7,8,2,3,4,10,11}
set1.intersection(set2)
print(set1)
```

# difference() → gives a new set (original not changed)
# difference_update() → updates original set by removing common elements

```python
set1={1,2,3,4,5,6}
set2={7,8,2,3,4}
set1.difference(set2)
```

```python
set1={1,2,3,4,5,6}
set2={7,8,2,3,4}
set1.difference_update(set2) #Removes elements found in another set.
print(set1)
```

```python
set1={1,2,3,4,5,6}
set2={7,8,2,3,4}
set1.symmetric_difference_update(set2) #Keeps only elements not common in both sets.
```

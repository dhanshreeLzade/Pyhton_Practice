```python
#Use lambda with map() to add 5 to each element in a set.
set1={1,2,3}
var=set(map(lambda a:a+5,set1))
print(var)
```

```python
#Use lambda with filter() to get numbers divisible by 3 from a list.
list1=[1,2,3,4,5,6,7,8,90]
var= list(filter(lambda a:a%3==0,list1))
print(var)
```

```python
# Use lambda with map() to find the length of each word in a tuple of strings.
tuple1=("jullyiooo","kullyioooooo","trully","fu","momo")
var=tuple(map(lambda a:len(a),tuple1))
print(var)
```

```python
#Use lambda with filter() to remove empty strings from a list.
list1=["jullyiooo","kullyioooooo","trully","fu","momo","",""]
var=list(filter(lambda a:a!="",list1))
print(var)
```

```python
#Use lambda with map() to capitalize dictionary keys.
dic1={'apple': 'APPLE', 'banana': 'BANANA'}
var=dict(map(lambda a:(a.capitalize(),dic1[a]),dic1))
print(var)
```

```python
#Use lambda with filter() to select values greater than 50 from a list.
list1=[1,2,312,232,221,12,34,56,65]
var=list(filter(lambda a:a>50,list1))
print(var)
```

```python
#Use lambda with map() to convert a list of strings into integers.
list1=["1","2","3","4","5"]
var=list(map(lambda a:int(a),list1))
print(var)
```

```python
# Use lambda with filter() to get names starting with vowel from a list.
list1=["aullyiooo","iullyioooooo","trully","fu","momo","",""]
var=list(filter(lambda a:a.startswith(('a','e','i','o','u','A','E','I','O','U')),list1))
print(var)
```

```python
#Use lambda with reduce() to find the product of numbers in a list.
list1=[1,2,3,4,5,6,7]
var=reduce(lambda a,b:a*b)
print(var)
```

```python
#Use lambda with reduce() to find the product of numbers in a list.
from functools import reduce
list1=[1,2,3,4,5,6,7]
var=reduce(lambda a,b:a*b,list1)
print(var)
```

```python
# Use lambda with map() to reverse each string in a tuple.
tup=("ananya","knanya","linanya","omnanya",)
var=tuple(map(lambda a:a[::-1],tup))
print(var)
```

```python
‚≠ê Easy examples:
‚úî Prime numbers:

2 ‚Üí can be divided only by 1 and 2

3 ‚Üí can be divided only by 1 and 3

5 ‚Üí can be divided only by 1 and 5

7 ‚Üí can be divided only by 1 and 7

üëâ These numbers do not divide evenly by any other number.

‚úñ Not prime (composite numbers):

4 ‚Üí can be divided by 1, 2, 4

6 ‚Üí can be divided by 1, 2, 3, 6

8 ‚Üí can be divided by 1, 2, 4, 8

üëâ These numbers divide evenly by many numbers.
```
